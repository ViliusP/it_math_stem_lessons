## Vizualizacija 

Importuojame vizualizacijai reikalingas bibliotekas:

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

Importuojame duomenis ir pažiūrime 5 pirmąsias eilutes:

```python
weather_data = pd.read_csv("./weatherAUS.csv")

weather_data.head()
```

Peržiūrime duomenų informaciją:

```python
weather_data.info()
```

```python
weather_numerics = weather_data.select_dtypes(include=[np.number])

# Now calculate the correlation matrix for the numeric data
corrmat = weather_numerics.corr()

# Continue with your plotting
cmap = sns.diverging_palette(260, -10, s=50, l=75, n=6, as_cmap=True)
plt.subplots(figsize=(18, 18))
sns.heatmap(corrmat, cmap=cmap, annot=True, square=True)
plt.show()
```
